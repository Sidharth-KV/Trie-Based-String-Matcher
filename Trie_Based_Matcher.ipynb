```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#define trie node
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

#trie class - for structure
class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for ch in word:
            if ch not in node.children:
                node.children[ch] = TrieNode()
            node = node.children[ch]
        node.is_end = True

    def get_prefix_path(self, prefix):
        node = self.root
        path = []
        for ch in prefix:
            if ch in node.children:
                path.append(ch)
                node = node.children[ch]
            else:
                break
        return path

    def print_trie(self, node = None, prefix = ""):
        if node is None:
            node = self.root
            print("Root")

        for ch, child in sorted(node.children.items()):
            new_prefix = prefix + ch
            if child.is_end:
                print(new_prefix)

#contacts of people as example
contacts = ["Alice", "Ali", "Alina", "Bob", "Brian", "Josh", "Ahmed", "Jesus", "Ram"]
trie = Trie()
for name in contacts:
    trie.insert(name)

# Visualize trie with networkx module
def build_graph(node, graph, current_label="root", prefix_path=[], path_so_far=""):
    for ch, child in node.children.items():
        new_label = f"{path_so_far}{ch}"
        graph.add_edge(current_label, new_label)

        # Color nodes in prefix path
        if new_label in prefix_path:
            graph.nodes[new_label]['color'] = 'lightgreen'
        else:
            graph.nodes[new_label]['color'] = 'lightblue'

        build_graph(child, graph, new_label, prefix_path, new_label)

def visualize_trie(prefix="Ali"):
    G = nx.DiGraph()
    G.add_node("root", color="lightgreen")

    path = trie.get_prefix_path(prefix)
    prefix_path_labels = []
    label = ""
    for ch in path:
        label += ch
        prefix_path_labels.append(label)

    build_graph(trie.root, G, "root", prefix_path_labels)

    pos = nx.spring_layout(G)
    node_colors = [G.nodes[n].get('color', 'lightblue') for n in G.nodes]

    plt.figure(figsize=(12, 8))
    nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=1200, font_size=10, font_weight='bold')
    plt.title(f"Trie Visualization (Prefix: '{prefix}')", fontsize=14)
    plt.show()

# Call the function with default prefix as parameter
visualize_trie("Ali")
```
